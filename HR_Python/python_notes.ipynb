##  3- Sets
True and 1 are considered the same value in sets, and are treated as duplicates.

To initiate a set: **myset=set()** <br/>
To sort a set: **sorted(myset)** <br/>

These operations do not make any changes or mutations to the set. <br/>
**x.union(y)** # Values which exist in x or y <br/>
**x.intersection(y)** # Values which exist in x and y **works only on sets** <br/>
**x.difference(y)** # Values which exist in x but not in y <br/>
**x.symmetric_difference(y)** keep only the elements that are NOT present in both sets. <br/>

These operations update the set: <br/>
**x.update(y)** adding elements from an iterable/another set.
**x.intersection_update(y)** update the set by keeping only the items that are present in both sets (x and y) <br/>
**x.difference_update(y)** Update the set by removing values found in y <br/>
**x.symmetric_difference_update(y)** Update the set by only keeping the elements found in either set, but not in both. <br/>

**x.issubset(y)** Returns whether another set contains this set or not (T/F) <br/>
**x.issuperset(y)** Returns True if x contains all elements of set A and 1 more element. <br/>

# Counter & Dictionnary
A Counter is a container that keeps track of how many times equivalent values are added. 

```python
# import counter class from collections module
from collections import Counter
 
#Creating a Counter class object using list as an iterable data container
a = [12, 3, 4, 3, 5, 11, 12, 6, 7]
 
x = Counter(a)
print(x)

#We can also use .keys() and .values() methods to access Counter class object
for i in x.keys():
      print(i, ":", x[i])
```

```python
#We can also make sets of keys and values of x
x_keys = set(x.keys())
x_values = set(x.values())
 
print(x_keys)
print(x_values)
```

### Parameters vs Arguments 
The terms parameter and argument can be used for the same thing: information that are passed into a function.
A parameter is the variable listed inside the parentheses in the function definition. <br/>
An argument is the value that is sent to the function when it is called.

### *args 
If you do not know how many arguments that will be passed into your function, add a * <br/>
Star operator allows you to pass unlimited number of arguments

```python
def print_names(name):
    print (name)

print_names('Mir', 'Rob') 
```

```python
def print_names(*name):
    print (name)

print_names('Mir', 'Rob')
```

```python
# star operator -> unpacking operator, allows you to print all items in the list 
# in a single print statement without using loops
print(*'Mireille')
print(*'Had, Christy')
items = [0,1,2,3,4]
print(*items) 
```

### Arbitrary Keyword Arguments, **kwargs
If the number of keyword arguments is unknown, add a double ** before the parameter name:

```python
def my_function(**kid):
  print("His last name is " + kid["lname"])

my_function(fname = "Tobias", lname = "Refsnes")
```


### Default argument
By default, a function must be called with the correct number of arguments.  But we can add a default value in the function to be used in case this value wasn't provided

```python
def to_smash(total_candies,n=3):
   
    return total_candies % n

print(to_smash(407,2))
print(to_smash(407))  # n is assumed to be 3
```

# Lamda
A lambda function can take any number of arguments, but can only have one expression.

```python
# With 1 argument
x = lambda a: a + 10
print(x(5))
```

```python
# with 2 arguments
x = lambda a, b : a * b
print(x(5, 6))
```

**The power of lambda** is better shown when you use them as an anonymous function inside another function. <br/>
Say you have a function definition that takes one argument, and that argument will be multiplied with an unknown number:

```python
def myfunc(n):
    return lambda x: x*n

# use the func to double any number
double = myfunc(2)
print(double(5))

# use the func to triple any number
triple = myfunc(3)
print(triple(5))
```

# Collections 
### collections.deque()

```python
from collections import deque
d = deque()
d.append(1)   # Add to the right side of the deque.
print (d)
```

```python
d.appendleft(2)  # Add to the left side of the deque.
print (d)
```

```python
d.clear() # Remove all elements from the deque leaving it with length 0.
print(d)
```

```python
d.extend('1')
print(d)
```

```python
d.extendleft('234') # Extend the left side of the deque by appending elements from iterable. they will be added 4, 3, 2 (reversed)
d.count('1')  # Count the number of deque elements equal to 1
d.pop() # Remove and return an element from the right side of the deque. If no elements are present, raises an IndexError.
d.popleft() #Remove and return an element from the left side of the deque. If no elements are present, raises an IndexError.
d.extend('7896') #Extend the right side of the deque by appending elements from the iterable argument.
d.remove('2') # Remove the first occurrence of value. If not found, raises a ValueError.
d.reverse() # Reverse the elements of the deque in-place and then return None.
print(d)
d.rotate(3) # Rotate the deque n steps to the right. If n is negative, rotate to the left.
print(d)
```

# re
A regular expression (or RegEx) specifies a set of strings that matches it.

A regex is a sequence of characters that defines a search pattern, mainly for the use of string pattern matching.


The **re.search()** expression scans through a string looking for the first location where the regex pattern produces a match.
It either returns a MatchObject instance or returns None if no position in the string matches the pattern.

```python
 import re
print (bool(re.search(r"ly","similarly")))
```

The **re.match()** expression only matches at the beginning of the string.
It either returns a MatchObject instance or returns None if the string does not match the pattern.

```python
print (bool(re.match(r"ly","similarly")))
print (bool(re.match(r"ly","ly should be in the beginning")))
```

The **re.split()** expression splits the string by occurrence of a pattern.

```python
re.split(r"-","+91-011-2711-1111")
```

# enumerate()
When you use enumerate(), the function gives you back two loop variables:

- The count of the current iteration
- The value of the item at the current iteration

```python
# without using enumerate()
values = ["a", "b", "c"]
for index in range(len(values)):
     value = values[index]
     print(index, value)
```

```python
# With enumerate()
for count, value in enumerate(values):
     print(count, value)
```
