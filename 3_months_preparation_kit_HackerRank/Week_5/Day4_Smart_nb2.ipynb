```python
import math

def is_smart_number(num):
    val = int(math.sqrt(num))
    # this line should be changed from == 1 to == val 
    if num / val == val:
        return True
    return False

for _ in range(int(input())):
    num = int(input())
    ans = is_smart_number(num)
    if ans:
        print("YES")
    else:
        print("NO")



```

## NB:
All positive integers have a paired number of factors, except for perfect squares which have an odd number of factors due to the pairing of identical factors. 

Perfect Squares: Perfect squares are numbers that are the result of multiplying an integer by itself. For example, 4, 9, 16, etc. When you factor a perfect square, you get pairs of identical factors. For example, the factors of 16 are 1, 2, 4, 8, and 16. Here, 2 and 8 are a pair, so perfect squares have an odd number of factors.
