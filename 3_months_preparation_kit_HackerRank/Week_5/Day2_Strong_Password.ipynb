```python
# !/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'minimumNumber' function below.
#
# The function is expected to return an INTEGER.
# The function accepts following parameters:
#  1. INTEGER n
#  2. STRING password
#

def minimumNumber(n, password):
    # Return the minimum number of characters to make the password strong
    numbers = "0123456789"
    lower_case = "abcdefghijklmnopqrstuvwxyz"
    upper_case = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    special_characters = "!@#$%^&*()-+"
    
    chr_change =0
    
    is_numbers = False
    is_lower_case = False
    is_upper_case = False
    is_special_characters = False

  
    for let in password:
        if let in numbers:
            is_numbers = True 
        elif let in lower_case:
            is_lower_case = True 
        elif let in upper_case:
            is_upper_case = True
        elif let in special_characters:
            is_special_characters = True
            
    
    # Find the nb of ch to change
    chr_change = 4 - (is_numbers + is_lower_case + is_upper_case + is_special_characters)
            
    if len(password)>=6 or (len(password)+ chr_change) >= 6:
        return(chr_change)
    else:
        return (6-len(password))
    
    

if __name__ == '__main__':
    #fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input().strip())

    password = input()

    answer = minimumNumber(n, password)
    
    print(answer)

    #fptr.write(str(answer) + '\n')

    #fptr.close()
```
