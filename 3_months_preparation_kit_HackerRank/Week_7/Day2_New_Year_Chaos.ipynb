```python
def minimumBribes(q):
    n = len(q)
    bribes = [0] * n  # Create a list to keep track of the number of bribes for each person

    for i in range(n):
        if q[i] - (i + 1) > 2:
            print("Too chaotic")
            return

        # Count the number of bribes for this person
        for j in range(max(0, q[i] - 2), i):
            if q[j] > q[i]:
                bribes[q[i] - 1] += 1

    total_bribes = sum(bribes)
    print(total_bribes)

if __name__ == '__main__':
    t = int(input().strip())

    for _ in range(t):
        n = int(input().strip())
        q = list(map(int, input().rstrip().split()))
        minimumBribes(q)
```
