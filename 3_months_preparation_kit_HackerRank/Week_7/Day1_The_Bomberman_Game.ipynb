```python
#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'bomberMan' function below.
#
# The function is expected to return a STRING_ARRAY.
# The function accepts following parameters:
#  1. INTEGER n
#  2. STRING_ARRAY grid
#

def getBombLocations(grid):
    # get initial bomb locations
    bomb_location = []
    
    for i in range(len(grid)):
        for j in range(len(grid[i])):
            if grid[i][j] == 'O': 
                bomb_location.append((i,j))
    return bomb_location

def getString(grid):
    return [''.join(x) for x in grid]
    
def bomberMan(n, grid):
    if n % 2 == 0:
        return [''.join(['O' for _ in range(len(grid[0]))]) for _ in range(len(grid))]
    
    states = [[list(row) for row in grid]]
    seen = [''.join(i for j in states[0] for i in j)]
    rows = len(grid)
    cols = len(grid[0])
    
    if n == 1:
        return getString(states[0])
    
    locations = getBombLocations(grid) # get initial bomb locations
    repeating = 0
    
    while True:
        full = [['O' for _ in range(len(grid[0]))] for _ in range(len(grid))]
        
        for e in locations:
            full[e[0]][e[1]] = '.'
            if e[0]-1 >= 0:
                full[e[0]-1][e[1]] = '.'
            if e[0]+1 < rows:
                full[e[0]+1][e[1]] = '.'
            if e[1]-1 >= 0:
                full[e[0]][e[1]-1] = '.'
            if e[1]+1 < cols:
                full[e[0]][e[1]+1] = '.'
    
        full_string = ''.join(i for j in full for i in j)
        
        try:
            repeating = seen.index(full_string)
            break
        except:
            pass
        
        seen.append(full_string)
        states.append(full)
        locations = getBombLocations(full)
    
    for i in range(repeating, len(states)):
        time = (i * 2) + 1
        print(i, len(states), time, n, len(states[repeating:]))
        if (n - time) % (2 * len(states[repeating:])) == 0:
            return getString(states[i])
        
        

if __name__ == '__main__':
    #fptr = open(os.environ['OUTPUT_PATH'], 'w')

    first_multiple_input = input().rstrip().split()

    r = int(first_multiple_input[0])

    c = int(first_multiple_input[1])

    n = int(first_multiple_input[2])

    grid = []

    for _ in range(r):
        grid_item = input()
        grid.append(grid_item)

    result = bomberMan(n, grid)
    
    print(result)

    #fptr.write('\n'.join(result))
    #fptr.write('\n')

    #fptr.close()
```
