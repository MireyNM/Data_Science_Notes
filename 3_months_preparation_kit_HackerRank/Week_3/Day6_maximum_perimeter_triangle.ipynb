### Solution that solves all cases

```python
def maximumPerimeterTriangle(sticks):
    sticks.sort()  # Sort the array in non-decreasing order

    # Iterate from the end of the array
    for i in range(len(sticks) - 1, 1, -1):
        # Check if a non-degenerate triangle is possible
        if sticks[i] < sticks[i - 1] + sticks[i - 2]:
            return [sticks[i - 2], sticks[i - 1], sticks[i]]  # Return sides in non-decreasing order

    return [-1]  # No non-degenerate triangle found

# Example usage:
stick_lengths = [1, 2, 3, 4, 5, 6, 7, 8, 9]
result = maximumPerimeterTriangle(stick_lengths)
print(result)  # Output: [7, 8, 9]
```
