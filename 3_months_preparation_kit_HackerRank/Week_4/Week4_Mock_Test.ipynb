```python
#!/bin/python3

import math
import os
import random
import re
import sys



#
# Complete the 'anagram' function below.
#
# The function is expected to return an INTEGER.
# The function accepts STRING s as parameter.
#

def anagram(s):
    # Write your code here
    # Split the string into 2 
    len_s = len(s)
    mid= int(len(s)/2)
    s_0 = (s[:mid])
    s_1 = (s[mid:])

    
    nb_change = 0
    for i in s_0:
        if i in s_1:
            continue
        else:
            nb_change +=1
    
    if len(s) % 2 != 0:
        return -1
    else:
        return nb_change
    

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    q = int(input().strip())

    for q_itr in range(q):
        s = input()

        result = anagram(s)

        fptr.write(str(result) + '\n')

    fptr.close()
```
