```python
def is_palindrome(s):
    return s == s[::-1]

def find_index_to_make_palindrome(word):
    if is_palindrome(word):
        return -1  # The word is already a palindrome

    for i in range(len(word)):
        modified_word = word[:i] + word[i+1:]
        if is_palindrome(modified_word):
            return i  # Index to remove for a palindrome

    return -1  # Cannot make a palindrome by removing any character

# Test cases
word1 = "aaab"
word2 = "baa"
word3 = "aaa"

print(find_index_to_make_palindrome(word1))  # Output: -1 (Already a palindrome)
print(find_index_to_make_palindrome(word2))  # Output: 2 (Remove 'l' to get "helo")
print(find_index_to_make_palindrome(word3))  # Output: -1 (Already a palindrome)
```
