```python
def flippingBits(n):
    # Write your code here
    x = list(format(n, '032b')) 
    for i in range(len(x)): 
        if x[i] == '0': 
            x[i] = '1' 
        elif x[i] == '1': 
            x[i] = '0' 
    print (int("".join(str(y) for y in x), 2))

if __name__ == '__main__':
    #fptr = open(os.environ['OUTPUT_PATH'], 'w')

    q = int(input().strip())

    for q_itr in range(q):
        n = int(input().strip())

        result = flippingBits(n)

        #fptr.write(str(result) + '\n')

    #fptr.close()
```

# Code Explanation 

## Convert the integer n into a binary string
- **format()** function is used to convert the integer n into a binary string representation. 
- **The format(n, '032b')** expression will return a binary string representing the integer n, formatted with leading zeros to make it 32 characters long.
- **list()** function is then used to convert this binary string into a list,

```python
n = 2147483647
x = list(format(n, '032b')) 
print(x)
```

## Iterate throuh each element of the list and flip it

```python
for i in range(len(x)):
    if x[i] == '0': 
        x[i] = '1' 
    elif x[i] == '1': 
        x[i] = '0' 
print(x)
```

## Converts a list of binary digits into its corresponding integer value.
- **int()** function is then used to convert this binary string into an integer. 
- The second argument of int() is 2, which specifies that the string represents a binary number. 
- The int() function interprets the string as a binary number and converts it to its equivalent integer value.

```python
print( int("".join(str(y) for y in x), 2))
```
