# What's Data Science?




### Data scientist 
Is someone who can extract meaning from complex data and assist in decision-making. "Someone who uses data in novel ways to build things.

### Data Science 
Is the study of data to extract meaningful insights for business.
It's a field that combines technical skills, analytical thinking, and domain expertise to extract valuable insights and knowledge from data to solve complex problems and make informed decisions.


# What's the difference between supervised and unsupervised learning?

### Supervised Learning
Supervised learning involves training a model on **labeled data** (prediction based on previous data) </br>
The desired output or target variable is known during the training process. </br>
Examples of supervised learning algorithms:
- Linear regression 
- Logistic regression
- Decision trees
- Support vector machines (SVM)
- Neural networks 

Supervised learning is commonly used for tasks such as classification and regression.  </br> 
**Classification** To predict continuous variables (weight) </br>
**Classification** To predict discretes outcomes (Yes/No)


### Unsupervised Learning
Unsupervised learning deals with unlabeled data, where the input data is not accompanied by any target labels or desired outputs. (Ask a bag to group objects: it might group them by shapes, colors or both...) </br>
Unsupervised learning algorithms aim to find hidden patterns or clusters in the data, identify anomalies, or reduce the dimensionality of the data. </br>
Examples of unsupervised learning algorithms:  
- Clustering algorithms like k-means, hierarchical clustering, and density-based clustering
- Dimensionality reduction techniques like principal component analysis (PCA) and t-distributed stochastic neighbor embedding (t-SNE).
