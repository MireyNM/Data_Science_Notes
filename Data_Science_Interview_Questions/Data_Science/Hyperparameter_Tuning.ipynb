# Hyperparameter tuning / hyperparameter optimization

### It is the process of finding the best combination of hyperparameters for a machine learning model to achieve optimal performance on a given dataset.

## Hyperparameters
Are parameters that are set before the learning process begins and cannot be learned directly from the training data. They control the behavior of the learning algorithm and affect the model's learning process and generalization capabilities.

If you ask me what is Hyperparameters in simple words, the one-word answer is Configuration.

Example: Hyperparameter is Train-Test Split Ratio (80-20) in our simple linear regression model.


## Data Leakage / Information Leakage
After successful testing with perfect accuracy, you get poor score/results when applying real-time/actual data. 


Occurs when information from outside the training dataset is used to create or influence a machine learning model, leading to overly optimistic performance results and compromised generalization ability.We should avoid doing anything to our training dataset that involves having knowledge of the test dataset. 
 

## Steps to Perform Hyperparameter Tuning
- Select the right type of model.
- Review the list of parameters of the model and build the HP space
- Finding the methods for searching the hyperparameter space
- Applying the cross-validation scheme approach
- Assess the model score to evaluate the model
