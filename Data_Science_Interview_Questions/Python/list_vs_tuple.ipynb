# What's the difference between lists and tuples

```python
my_list = [1, 2, 3]  # List
my_tuple = (1, 2, 3)  # Tuple

# Modifying elements
my_list[0] = 4  # List can be modified
# my_tuple[0] = 4  # Tuple throws an error since it is immutable

# Adding elements
my_list.append(4)  # Adding an element to the list is possible
# my_tuple.append(4)  # Tuples do not have an append method

# Memory and performance
import sys
print(sys.getsizeof(my_list))  # Memory size of list
print(sys.getsizeof(my_tuple))  # Memory size of tuple
```

### In python, we can use tuple objects as keys for python dictionaries. 
But we can not use list objects as keys for dictionaries. Because, keys must be immutable and hashable.

In this example, a dictionary named student_grades is created, where the keys are tuples representing the names of the students

```python
student_grades = {
    ('John', 'Smith'): 85,
    ('Jane', 'Doe'): 92,
    ('Michael', 'Johnson'): 78
}

# Accessing values using tuple keys
print(student_grades[('John', 'Smith')])  # Output: 85

# Iterating over dictionary with tuple keys
for student_name in student_grades:
    print(student_name, ":", student_grades[student_name])
```
