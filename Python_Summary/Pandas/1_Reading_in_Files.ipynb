### We need to add a seperator

sep='\t' tells pd.read_csv to interpret tabs (\t) as column delimiters in the text file, meaning each tab-separated entry will be treated as a new column.

```python
df = pd.read_csv('Data/countries of the world.txt', sep = '\t')
df
```

### .loc accessor: Label-based indexing
- Purpose: Access rows and columns using labels or names.
- Syntax: df.loc[row_labels, column_labels]
- Inclusive of End Points: When using ranges, .loc is inclusive of both the start and end labels.

```python
# Select row with label 2 and column 'Country'
df2.loc[2, 'Country']
```

```python
# Select rows labeled 0 to 3 (inclusive) and columns 'Country' and 'Region'
df2.loc[0:3, ['Country', 'Region']]
```

```python
# Select rows where 'Region' is 'Brazil'
df2.loc[df2['Region'] == 'Brazil']
```

### .iloc: Integer-based indexing
- Purpose: Access rows and columns using integer positions.
- Syntax: df.iloc[row_indices, column_indices]
- Exclusive of End Points: When using ranges, .iloc excludes the end index.


```python
#Select the data at the third row (index 2) and second column (index 1)
df2.iloc[2, 1]
```

```python
# Select the first three rows and first two columns
df.iloc[0:3, 0:2]
```
