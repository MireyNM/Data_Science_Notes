## 1. Line Plot

```python
df.plot(kind = 'line',
        title = 'Ice Cream Ratings', 
        xlabel = 'Daily Ratings', 
        ylabel = 'Scores')
```

## 2. Bar Plot

```python
df.plot(kind = 'bar') 
```

### 2.1 Stacked Bar Plot

```python
df.plot(kind = 'bar',stacked = True)
```

### 2.2 Stacked Horizontal Bar Plot
Used:
- To compare totals across categories and breakdowns within those totals.
- To visualize data where labels or categories are easier to read horizontally (e.g., long group names).

```python
df.plot.barh(stacked = True)
```

## 3. Scatter plots
Used when:
- You're analyzing the relationship between two numerical variables.
- You want to explore correlation, trends, or outliers.
- You need a visual representation of individual data points.

When Not to Use a Scatter Plot:
- When comparing categorical data: Use bar plots, box plots, or pie charts instead.
- When visualizing more than two variables: Consider 3D scatter plots or alternative methods like pair plots, heatmaps, etc.
- When summarizing data trends: Line plots or histograms may be better for summarizing trends.


```python
df.plot.scatter(x = 'Texture Rating',
                y = 'Overall Rating', 
                s = 500, 
                c = 'Yellow')
```

## 4. Histogram plot
Use a histogram when:
- You want to analyze the distribution of a single numerical variable.
- You're exploring patterns, such as symmetry, skewness, or multimodality.
- You need to identify variability or outliers in continuous data.

When Not to Use a Histogram:
- Categorical Data: Use bar plots instead for non-numerical data.
- Small Datasets: If the dataset has very few values, histograms might not be meaningful.

```python
df.plot.hist(bins = 10)
```

## 5. Box plot

What the Box Plot Shows:
- Median (Line inside the box): The middle value of the data.
- Box: Represents the interquartile range (IQR = Q3 - Q1), the middle 50% of the data.
- Whiskers: Extend to the smallest and largest data points within 1.5 times the IQR.
- Outliers: Data points outside the whiskers are plotted individually.

Use Cases:
- Exploratory Data Analysis (EDA): Quickly assess distributions and outliers.
- Comparing Groups: Analyze distributions across categories (e.g., age groups, test scores).
- Data Quality Checks: Spot outliers or anomalies in the data.

When Not to Use a Box Plot:
- Small Datasets: If the dataset is too small, box plots might not provide meaningful insights.
Use raw data points or strip plots instead.
- Categorical Data: For non-numerical variables, box plots are not suitable.

```python
df.boxplot()
```

## 6. Area Plot
Use Cases:
- Time Series Data: Visualizing how values accumulate or change over time.
- Comparing Contributions: Understanding how different components contribute to an overall total.
- Showing Cumulative Data: Tracking a growing or accumulating quantity, like total population, revenue, or sales.

When Not to Use an Area Plot:
- When Data is Too Noisy: If the data has large fluctuations or if individual data points are too close to each other, an area plot might become difficult to interpret.
Example: Financial data with high volatility could become hard to interpret in an area plot.
- When Comparing a Single Category or Variable: If youâ€™re only concerned with one variable or trend, a line plot or bar plot may be more effective.
- Too Many Variables: If you have too many categories or variables, the plot can become cluttered, making it harder to see individual trends or contributions.

```python
df.plot.area(figsize = (10,5))
```

## 7. Pie Chart
Use a pie chart when:

- You want to show proportions or percentages of categories in a whole.
- You have a small number of categories (2-6).
- You want a simple, intuitive representation of data.

```python
df.plot.pie(y='Flavor Rating',figsize=(10,10))
```
