```python
import pandas as pd
```

```python
df = pd.read_csv('Data/world_population.csv')
df.head()
```

## Check Column names

```python
print(df.columns)
```

## Check for Extra Spaces in the columns names and strip them
.str.strip() will remove any leading or trailing whitespace from the column names

```python
df.columns = df.columns.str.strip()
```

## Filter all rows with a column value bigger than X

```python
## Output all rows with rank value bigger than 100
df[df['Rank']>100]
```

## isin() 
df[df['column_name'].isin(list_of_values)]

Purpose: Filters rows by checking if column values match any item in a specified list, set, or iterable.

```python
specific_countries = ['Andorra', 'Brazil']
andorra_brazil = df[df['Country'].isin(specific_countries)]
andorra_brazil
```

## .str.countains()

Purpose: Checks if a substring exists within each value in a column of strings. It’s useful for partial matches.

```python
df[df['Country'].str.contains('United')]
```

## .set_index()

Used to set one or more columns of a DataFrame as the index

df.set_index(keys, drop=True, inplace=False)

- keys: The column (or columns) to set as the index.
- drop: Whether to drop the column(s) from the DataFrame after setting it as the index. Default is True.
- inplace: If True, modifies the original DataFrame. If False, returns a new DataFrame with the index set

```python
df2 = df.set_index('Country')
df2
```

## .filter() 
To select specific columns or rows from the DataFrame

axis=1: Specifies that the filtering is applied to columns. 

In Pandas, axis=1 refers to columns, while axis=0 would refer to rows.

```python
# Select specific columns from the DataFrame
df2.filter(items = ['Continent','CCA3'], axis = 1)
```

```python
# Select specific rows from the DataFrame
df2.filter(items = ['Zimbabwe'], axis = 0)
```

### .filter(like=...)
To select rows or columns with labels containing specific text

```python
df2.filter(like = 'United', axis = 0)
```

## .loc()

```python
df2.loc['United States']
```

## .iloc()

```python
df2.iloc[3]
```

## Order by

```python
df[df['Rank'] < 10].sort_values(by=['Continent','Country'],ascending=[False,True])
```
