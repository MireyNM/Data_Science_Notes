# Why do we use CASE Statement in SQL? Give example.
CASE statement is similar to IF ELSE statement from any other programming languages. We can use it to fetch or show a particular value based on certain condition. </br>

Here's an example to illustrate the use of the CASE statement: </br>

Suppose you have a table named "students" with columns "name," "score," and you want to categorize students' performance as "Excellent," "Good," "Average," or "Poor" based on their scores.



In this example:

The CASE statement is used to evaluate different conditions:
- If a student's score is 90 or higher, they are categorized as "Excellent."
- If the score is between 80 and 89, they are categorized as "Good," and so on.
- The ELSE part covers any score that doesn't fit into the previous conditions, and those students are categorized as "Poor."
- The result of the CASE statement is aliased as "performance" in the output.
