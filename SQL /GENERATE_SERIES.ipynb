# GENERATE_SERIES
## A function in SQL that generates a series of values within a specified range. It's commonly used to create sequences of numbers or dates



### SELECT * FROM GENERATE_SERIES(1, 10); 

Will generate a list of number from 1 to 10 (Each one on a row)

### SELECT * FROM GENERATE_SERIES('2023-08-01'::date, '2023-08-07'::date, '1 day');
This query generates a sequence of dates from August 1, 2023, to August 7, 2023, with a step size of 1 day.

### SELECT * FROM GENERATE_SERIES(0, 100, 10);

This query generates a sequence of numbers from 0 to 100, with a step size of 10.
