# Define a unique identifier
A unique identifier is a unique value that identifies a record so that it can be distinguished from other records in the same table.

# Data Types 
1. Numeric Data Types:
   - INT or INTEGER: Integer values (whole numbers). (-2 billion to + 2 billion)
   - SMALLINT: Small integer values.
   - BIGINT: Large integer values.
   - DECIMAL or NUMERIC: Fixed-point decimal numbers.
   - FLOAT or REAL: Floating-point numbers.
   - NUMERIC: store floats up to 38 digits total includin those before and after the decimal point
   - DOUBLE or DOUBLE PRECISION: Double-precision floating-point numbers.

2. Character Data Types/Strings:
   - CHAR: Fixed-length character string. Up to 250 charachter. Ex: CHAR(10) indicates a string of 10 characters or shorter. (Save storage space)
   - VARCHAR: Variable-length character string (up tp 10,000 charachters)
   - TEXT: Variable-length character string for larger text values.

3. Date and Time Data Types:
   - DATE: Date values (year, month, day).
   - TIME: Time values (hour, minute, second).
   - DATETIME or TIMESTAMP: Combination of date and time values.
   - INTERVAL: Time interval or duration.

4. Boolean Data Type:
   - BOOLEAN or BOOL: Represents true/false or logical values.

5. Binary Data Types:
   - BINARY: Fixed-length binary data.
   - VARBINARY: Variable-length binary data.
   - BLOB: Binary large object for storing larger binary data.

6. Miscellaneous Data Types:
   - ENUM: Represents a set of predefined values.
   - JSON: Stores JSON (JavaScript Object Notation) data.
   - UUID: Universally unique identifier.



# Schemas 
**Blueprints of databases** A schema shows a database's design, such as what tables are included in the database and any relationships between its tables. A schema also lets the reader know what data type each field can hold
