# What are DDL and DML languages? Give example.

SQL can be divided into 2 categories: 

1 - DDL </br>
2 - DML </br>
3 - DQL 

## DDL  = Data Definition Language. 
DDL statements are used to define the structure of a database and its objects. They are used for creating, altering, and dropping database objects like tables, indexes, views, and more. DDL statements do not manipulate the data within the tables; they focus on defining the schema and layout of the database.

Common DDL statements include: 
- CREATE: Used to create database objects like tables, indexes, views, etc.
- DROP: Used to remove database objects from the schema.
- ALTER: Used to modify the structure of existing database objects. 
- TRUNCATE: used to quickly remove all rows from a table while keeping its structure intact


## DML = Data Manipulation Language
DML statements are used to manipulate the data stored within the database objects, primarily tables. They allow you to insert, retrieve, update, and delete data from the tables. DML statements do not alter the structure of the database; they focus on managing the actual data within it.

Common DML statements include:

- SELECT: When it's used in certain contexts that involve manipulating data,
- INSERT: Used to add new rows of data into a table.
- UPDATE: Used to modify existing data in a table.
- DELETE: Used to remove rows of data from a table.



In summary, DDL is focused on defining and managing the structure of the database and its objects, while DML is focused on manipulating the actual data stored within those objects.
